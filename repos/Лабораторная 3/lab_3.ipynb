```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
sys.path.append('..')
```

```python
image = cv.imread('image2.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image = cv.bitwise_not(image)

plt.imshow(image)
```

```python
gamma_corrected = np.array(255*(image / 255) ** 1.85, dtype = 'uint8')

plt.imshow(gamma_corrected)
```

```python
gaussian = cv.GaussianBlur(image,(5,5),1)

plt.imshow(gaussian)
```

```python
kernel = np.asarray([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

image_median = cv.medianBlur(gamma_corrected, 3)

filtered_image = cv.filter2D(image_median, -1, kernel)

plt.imshow(filtered_image)
```
