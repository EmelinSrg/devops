```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread('input.jpg', 0)
plt.imshow(image, cmap='gray')
```

```python
thresh = cv.adaptiveThreshold(image, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 35, 90)
plt.imshow(thresh, cmap='gray')
```

```python
kernel = np.ones((8, 4), np.uint8)
opening = cv.morphologyEx(thresh, cv.MORPH_OPEN, kernel)
plt.imshow(opening, cmap='gray')
```

```python
kernel = np.ones((6, 4), np.uint8)
erosion = cv.erode(thresh, kernel, iterations=1)
plt.imshow(erosion,cmap='gray')
```
